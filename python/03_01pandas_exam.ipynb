```python
import pandas as pd
```

```python
# 1. exam 데이터 불러오기
exam = pd.read_csv('/content/drive/MyDrive/data/exam.csv')
exam
```

```python
# 2. exam 데이터의 정보 확인해보기
exam.info()
```

```python
# 3. exam 데이터의 요약보기
exam.describe(include='all')
```

```python
#4. exam데이터의 각 평균값 구하기
#4.1 수학값 가져고 오기
exam['math']
#4.2 영어값 가지고 오기
exam['english']
#4.3  과학값 가지고 오기
exam['science']
```

```python
exam_mean = (exam['math']+exam['english']+exam['science'])/3
exam_mean
```

```python
#5. 평균 컬럼 파생변수 만들기
exam['avg'] = exam_mean
exam
```

```python
import numpy as np
```

```python
# 평균점수가 60점 이상인 사람 통과 pass, fail
# pass,  fail의 값도 확인

exam['통과'] = np.where(exam['avg']>60,'pass','fail')
exam
```
