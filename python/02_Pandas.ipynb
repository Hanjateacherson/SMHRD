#### Series

```python
#여러 값을 나열한 자료구조
# index - value
# dataframe구성하는 하위요서 ----> 여러개의 Series는 DataFrame이 됨
```

```python
import pandas as pd
```

```python
#. 값을 가진 series 만들기
# pd.Series(리스트)
num_series = pd.Series([1,2,3,4,5])
num_series

#인덱스 미 작성시 0~데이터 수만큼 자동생성
```

```python
#인덱스를 지정한 series 만들기

num_series2 = pd.Series([10,20,30], index = ['kim','lee','park'])
num_series2
```

```python
#인덱스 series 만들기 22
num_series3 = pd.Series({'kim': 10, 'lee' :20, 'park' : 30})
num_series3
```

```python
#value 값 확인
#.values
num_series3.values
```

```python
#index 값 확인
#.index
num_series3.index
```

```python
#값의 데이터 타입 확인
# .dtype
num_series3.dtype
```

```python
# 시리즈의 이름을 부여
#.name = 이름명
num_series3.name = 'series 예제'
num_series3
```

```python
#인덱스 이름 부여
#.index.name = 인덱스 이름명
num_series3.index.name = '성'
num_series3
```

```python
population = pd.Series([9602000,3344000,1488000,2419000], index = ['서울','부산','광주','대구'])
population.index.name = '도시'
population.name = '2020 인구수'
```

```python
population
```

```python
#시리즈 값 수정
#인덱싱 간단하게
population['부산']
```

```python
# 인덱싱 = 수정 값
population['부산'] =3500000
population
```

```python
#값 추가
population['대전'] = 1500000
population
```

```python
# 값 삭제
# 1. .drop(index명) -- 값을 다시 대입 하지 않습니다
#drop을 사용해서 해당 index 삭제하고 싶을시 대입까지 진행
population = population.drop('서울')

population
```

```python
#2. del 키워드 사용 - 별도의 대입 없이 삭제가 가능
del population['대전']

population
```

## 데이터 프레임

```python
#1개의 시리즈가진 데이터 프레임
# pd.DataFrame()

num_df =pd.DataFrame([1,2,3])
num_df
```

```python
#2중 리스트로 값을 여러개 넣는 것도 가능!
num2_df = pd.DataFrame([['iot',1],['big data', 2],['ai', 3]])
num2_df
```

```python
#컬럼 값 지정하기
num2_df = pd.DataFrame([['iot',1],['big data', 2],['ai', 3]], columns=['class','join'])
num2_df
```

```python
#컬럼 값 지정하기 2
num2_df = pd.DataFrame({'class' : ['iot','big data', 'ai'], 'join' : [1,2,3]})
num2_df
```

```python
#인덱스 값 지정하기
num2_df = pd.DataFrame({'class' : ['iot','big data', 'ai'], 'join' : [1,2,3]} , index = ['A','B','C'])
num2_df
```
