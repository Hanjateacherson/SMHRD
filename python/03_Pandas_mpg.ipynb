### 컬럼 -- grade (등급)
- 평균값이 25이상이면 A 등급
- 20이상 B 등급
- 그 외 C 등급

## apply()라는 함수
- np.where의 조건식 복잡해지는 것을 간단하게 작성 할 수 있도록 하는 함수


```python
# apply() 사용법

# 매개변수가 함수
# 함수를 만들어서 함수명만 작성 해주시면 됩니다!

def makeGrade(df): # 함수 실행시 조건이 될 시리즈가 필요! 매개변수로 df 작성
    grade = df['평균']

    if grade >= 25:
      return "A등급"
    elif grade >= 20:
      return "B등급"
    elif grade >= 15:
      return "C등급"
    else:
      return "F등급"
```

```python
# .apply()
mpg_copy['grade2'] = mpg_copy.apply(makeGrade,axis=1)
mpg_copy
```

```python
mpg_copy['grade2'].value_counts().sort_index()
```

```python
#agg() - 다중집계작업
#agg(결과이름, 통계함수)
#통계 함수 : sum, mean, max , min ,count
```

```python
# 제조사별 도시연비 평균
mpg_copy.groupby('제조사').agg(city_mean = ('도시','mean'))
```

```python
mpg_copy.groupby('제조사').agg(city_mean = ('도시','count'))
```

```python
# 여러개 사용도 가능!
# 도시평균, 고속도로 평균

#.groupby로 제조사별로 묶기
#.agg
mpg_copy.groupby('제조사').agg(city_mean = ('도시','mean'), high_mean= ('고속도로','mean'))
```

```python
# 카테고리별 도시연비
# 카테고리 조사
mpg_copy['category'].value_counts()
```

```python
# 해당 결과값 정렬
# error! ➡️ 'by' argument
# by - 어떤 컬럼을 바탕으로
mpg_copy.groupby('category').agg(city_mean= ('도시','mean')).sort_values(by = 'city_mean',ascending= False )
```

```python
# pd.merge()


# db - join inner, outer - left right
# left = df1이 기준
# right = df2가 기준
# outer - 데이터를 모두 합치는데 만약에 df1이나 df2와 일치하지 않는 데이터
#      ↪️ nan값이 들어가게 됨
# test3로 확인

# inner - 교집합 부분만 표시 (일치하지 않는 데이터는 merge ❌)

# pd.merge(df1, df2, how = '')
```

```python
test1 = pd.DataFrame({'id':[1,2,3,4,5,6],'midterm':[60,80,70,90,85,80]})
test2 = pd.DataFrame({'id':[1,2,3,4,5],'final':[70,83,65,95,80]})
```

```python
test3 = pd.merge(test1, test2, how = 'outer', on = 'id')
test3
```

```python
test4 = pd.merge(test1, test2, how = 'inner', on = 'id')
test4
```
