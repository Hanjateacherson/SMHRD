# numpy 인덱싱, 슬라이싱 - 데이터 접근 및 처리

```python
#1차원 배열 만들기
arr4 = np.array([10,20,30,40,50,60])
arr4
```

```python
#인덱싱
# 리스트의 인덱싱 - list명[인덱스 숫자]
# 넘파이의 배열도 동일
# 인덱싱 방법 --> []


print(arr4[3])

#한번에 여러요소를 접근하는 경우 대괄호 한번 더 쓰워주기
print(arr4[[0,4]]) #0번째 4번째
```

### 1차원 슬라이싱

```python
# 리스트 슬라이싱 - [:]
# 배열도 동일
print(arr4[1:5])
print(arr4[:5])
print(arr4[2:])
print(arr4[:]) #모두
```

### 2차원 인덱싱

```python
 # 1차원 배열을 2차원 배열로 만들어주는 함수
 # .reshape(행,열)
print(arr4)
arr5 = arr4.reshape(2,3)

print(arr5)
```

```python
print(arr5[0]) #10,20,30
print(arr5[0][1])

#2차원 배열 데이터 접근하는 방법 1가지 더 존재
#[행,열] - ,(콤마) 기준 앞이 행 뒤가 열
print(arr5[0,1]) #arr5[0][1]와 동일
```

#### 2차원 슬라이싱

```python
arr5[:]
```

```python

arr5[0,:] # 콤마를 기준으로 앞은 행 뒤는 열
#행은 0번째 (인덱싱), 열은 슬라이싱으로 전부 출력
```

```python
# 20, 30만 출력하고 싶다

print(arr5[0,1:])

# 40, 50
print(arr5[1,:2])

#10,20,40,50
print(arr5[0:2,0:2])
```

```python
# 모름

arr5[:,[0,2]] #전체행 ,[0번째,2번째]
```

```python
# 모름
arr5[:,::2] # [:,::step] step작성
```

```python
# 데이터 수정
# 수정하고 싶은 위치 접근 후 대입 -- 인덱싱(슬라이싱) 한 후 = 원하는 변수
# arr5의 0행에 있는 데이터를 모두 100으로 수정
arr5[0] =100
# arr5[0] == arr5[0,:]
arr5

print(arr5)

arr5[0, :] = [10,20,30]
print(arr5)
```

```python
#데이터 추가
# list의 경우  append와 insert
# append(추가할 array, 추가 할 값) - 맨끝에 추가
arr6 = np.append(arr5, 30) # 끝에 30이 추가
arr6
```

```python
# 데이터 병합
# concatenate
list1= [1,2,3,4,5]
list2 = [10,20,30,40,50]
print(list1+list2) #리스트의 겨우 더할 시 데이터가 병합의 느낌
```

```python
arr7 = np.array([1,2,3,4,5])
arr8 = np.array([10,20,30,40,50])

arr_concat = np.concatenate([arr7,arr8]) #안에 리스트형태로 들어가야한다
arr_concat
```

### 조건에 맞는 데이터 접근하기

#### 불리언인덱싱


```python
bool_arr = [True, False]

#numpy에 있는 random함수
np.random.seed(1) #1번 씨드로 고정
bool_arr1= np.random.choice(bool_arr, size=10)
# True, False의 값을 랜덤으로 10개 만들겠습니다

print(bool_arr1)

temp_arr = np.arange(10)
print(temp_arr)

#temp_arr 불리언 인덱싱
# 불리언 인덱싱 - True인 값들 만 출력
# 불리언 인덱싱 - True, False가 담긴 array를 인덱싱
temp_arr[bool_arr1] # True인 값들 만 출력
```

```python
# 작성한 조건으로 불리언 인덱싱

# 성적데이터를 램덤으로 생성
np.random.seed(7)
score_arr = np.random.randint(50,100, size=20)
display(score_arr)

student_arr = np.arange(1,21)
display(student_arr)
```

```python
#논리연산자 사용해서 80점 이상인 학생들만 출력
# <, >, ==, and, or, not, in
score_arr[score_arr >80]
```

```python
score_arr[score_arr >80].size #7명 #몇 명인지 알려준다
```

```python
# 성적이 60점 이상인 학생들을 출력
student_arr[score_arr>=60]
```

### 데이터 정보 확인하기

```python
print(rating.ndim)
# 차원 수 확인
print(rating.shape)
print(rating.size)
```

### 인덱싱/슬라이싱

```python
# 평점 데이터
rating[:,2]
```

```python
# 평점들의 평균 - mean()
rating_mean = rating[:,2].mean()
print(rating_mean)
```

```python
# 사용자 id만 접근
rating_id = rating[:,0]
print(rating_id)
```

```python
#unique() 함수 사용
# np.unique(배열명, return_counts =True) ⏩ 해당 중복된 값의 개수를 리
unique_id = np.unique(rating_id)
print(unique_id)
print(unique_id.shape)
```

```python
np.unique(rating_id, return_counts=True )
```

```python
# 불리언 인덱싱
# user_id가 1인 사람의 평균평점
# ↪️ 1. user_id 1인 사람 구하기
user_id1 =rating[rating[0:,0]==1]
print(user_id1.shape)
# ↪️ 2. 1인 사람의 평점
user_id1[:,2]
# ↪️ 3. 평점의 평균
user_id1[:2].mean()
```

```python
#user_id ==1
rating[rating[:,0]==1][:,2].mean()
```

```python
# user_id == 2
rating[rating[:,0]==2]
```

```python
# user_id == 3
rating[rating[:,0]==3][:,2].mean()
```

```python
user_id = np.unique(rating_id)
user_id
```

```python
rating_result = []
for i in user_id:
  rating_mean = rating[rating[:,0]==i][:,2].mean()
  rating_result.append([user_id, rating_mean])
#print(rating_mean)
```
